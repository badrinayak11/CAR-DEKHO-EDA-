```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import plotly.graph_objs as go
import plotly.express as px
```

```python
df = pd.read_csv(r'/content/CAR DETAILS FROM CAR DEKHO (1).csv')
```

```python
df.head(10)
df.tail(10)
df.isnull().sum()
df.describe()
df.info()
df.shape
```

```python
df.head(10)
```

```python
df["company_name"] = df.name.apply(lambda x : ' '.join(x.split()[:1]))
df.head(10)
```

```python
df.company_name.value_counts()
```

```python
sns.countplot(data= df, x= 'company_name',palette= "CMRmap")
plt.xticks(rotation= 90)
plt.xlabel('company_name')
plt.ylabel('count')
plt.title('counts of company_name')
plt.show()
```

```python
labels = df['company_name'][:30].value_counts().index
sizes = df['company_name'][:30].value_counts()
colors = ["green", "yellow","pink","orange","blue","purple","red"]
plt.figure(figsize=(8,8))
plt.pie(sizes, labels = labels, colors = colors, autopct='%.2f%%')
plt.title('company_name')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS

text = ' '.join(df['company_name'])

plt.rcParams['figure.figsize'] = (8,8)
wordcloud = WordCloud(background_color = 'black',colormap='vlag', width = 1000,  height = 1000, max_words = 121).generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
plt.show()
```
